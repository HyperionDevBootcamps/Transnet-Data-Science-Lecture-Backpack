# Data Visualisation

Data visualisation is an essential tool for making data more comprehensible, revealing patterns, trends, and relationships that might otherwise remain hidden. To figure out what kind of visualisation tool is appropriate, it's important to fisrt understand the variables you are working with.

We will demonstrate five common types of visualisations: line graphs, bar charts, pie charts, histograms, and scatterplots. With each visualiation, we will discuss what kinds of variables are in the dataset and why our chosen visualisation tool is appropriate.

## FMRI Dataset

This dataset records brain activity measured by functional MRI over time. It contains information about different timepoints during an experiment, the brain signal strength, and the type of event (stimulus or control) being observed.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset("fmri")

fmri.head()
```

### Line Graph

A line graph shows trends over time or another continuous variable.

```python
plt.figure(figsize=(5, 4))
sns.lineplot(x="timepoint", y="signal", data=fmri, ci = None)
plt.title('Line Graph: FMRI Signal over Time')
plt.show()
```

## Tips Dataset

This dataset contains information about the total bill, tip amount, and other customer-related details from a restaurant. It includes variables such as the day of the week, time of day, and the size of the dining party.

```python
tips = sns.load_dataset("tips")

tips.head()
```

### Bar Chart
A bar chart is used to display and compare discrete categories of data through rectangular bars, where the length or height of each bar is proportional to the frequency or value of the corresponding category.

```python
plt.figure(figsize=(5, 4))
sns.barplot(x="day", y="total_bill", data=tips, ci = None)
plt.title('Bar Chart: Total Bill by Day')
plt.show()
```

### Pie Chart

A pie chart shows proportions of a whole.

```python
# Count occurrences of each day
day_counts = tips['day'].value_counts()
day_counts
```

```python
# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title('Pie Chart: Distribution of Days')
plt.show()
```

## MPG Dataset

This dataset provides information on various car models, including miles per gallon (MPG), horsepower, weight, and origin of the car.

```python
mpg = sns.load_dataset("mpg")

mpg.head()
```

### Histogram

A histogram displays the distribution of a numeric variable.

```python
plt.figure(figsize=(8, 6))
sns.histplot(mpg['mpg'], bins=20)
plt.title('Histogram: Distribution of MPG')
plt.show()
```

## Iris Dataset

This dataset contains measurements of sepal length, sepal width, petal length, and petal width for three species of iris flowers. It's a classic dataset in machine learning!

```python
iris = sns.load_dataset("iris")

iris.head()
```

### Scatterplot

A scatterplot visualises the relationship between two numerical variables.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.title('Scatterplot: Sepal Length vs Sepal Width')
plt.show()
```
