### Dealing with Missing Values

For some rows, we do not have the number of pets recorded.

\\

The missing values for the number of pets seem to be completely random (MCAR), meaning theres no clear reason why theyre missing. Since its random, using the column average to fill in the blanks makes sense. This way, we keep the overall mean intact without messing up the data or adding bias that could affect certain groups more than others.

```python
df.isnull().sum()
```

```python
avg_num_pets = round(df['Num of pets in household'].mean(), 0)
avg_num_pets
```

```python
df = df.fillna(avg_num_pets)
df.isnull().sum()
```
